# Observations:
    
Happiness Score: This is the target variable if you’re trying to understand or predict happiness scores based on other factors.
    
Economy (GDP per Capita), Family, Health (Life Expectancy), Freedom, Trust (Government Corruption), Generosity: These are all potential predictor variables that could be correlated with the Happiness Score.

Dystopia Residual: This column represents the unexplained components for each country’s happiness score. It might not be necessary if you’re only interested in the relationship between happiness scores and the six factors listed above.

Standard Error: This column represents the standard error of the happiness score. It might not be necessary for most analyses, but could be useful if you’re interested in the reliability or variability of the happiness scores.

The Happiness Score in the World Happiness Report is indeed a composite measure that’s calculated as the sum of the scores of all the factors you mentioned. Therefore, using these same factors to predict the Happiness Score could lead to perfect multicollinearity, meaning that our predictor variables are highly correlated with each other. This can make our model unreliable and the estimates of the coefficients can be highly sensitive to small changes in the model.

Instead, a more reliable approach might be to use these factors to predict the residuals of the happiness scores after accounting for the Dystopia Residual. This would give us an idea of how much each factor contributes to the happiness score above or below what would be predicted by a dystopian benchmark.

# Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

```python
maxAccu = 0
maxRs = 0
for i in range(1,2000):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = i)
    lr = LinearRegression()
    lr.fit(x_train,y_train)
    pred = lr.predict(x_test)
    acc = r2_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i 
print('Best r2 score is =', maxAccu,'at random state =', maxRS )
```

```python
#Creating train and test split at random state = 260
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = maxRS)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.linear_model import Lasso,Ridge
import xgboost as xgb
```
